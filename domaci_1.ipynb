# Definisanje opštih funkcija

## Ciljna funkcija

### Definicija ciljne funkcije

```python
def fc(x,y):
    """
    Opis funkcije:
        Funkcija za ulazne argumente x i y vraća
        vrednost ciljne funkcije za date koordinate.
    Ulazni argumenti:
        x : Argument funkcije prilagođenosti
        y : Argument funkcije prilagođenosti
    Izlazni argument:
        f(x,y) : Vrednost ciljne funkcije
    """
    return 2 - 0.1*(np.cos(5*pi*x)+np.cos(5*pi*y))-x**2-y**2
```

### Prikaz ciljne funkcije

```python
x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)

X, Y = np.meshgrid(x, y)
Z = fc(X, Y)

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 200)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('Ciljna funkcija')
ax.view_init(10, 45)
plt.show()

fig.savefig('C:\Z\ETF\Vestacka inteligencija\DZ1\Izvestaj\ciljna_funkcija.png', bbox_inches='tight')
```

## Nasumičan izbor suseda

### Definisanje funkcije za nasumičan izbor suseda

```python
def generisanje_suseda(x, y):
    """
    Opis funkcije:
        Na osnovu koordinata trenutnog stanja, nasumično se
        bira red veličine pomeranja i množi se za brojem u
        opsegu (0,1). Dodavanjem pomeraja na koordinate koje
        su date kao ulazni argumenti, generišu se susedi.
    Ulazni argumenti:
        x : Argument funkcije prilagođenosti
        y : Argument funkcije prilagođenosti
    Izlazni argument:
        xp : x koordinata generisanog suseda
        yp : y koordinata generisanog suseda
    """
    
    # Moguća osnovna pomeranja suseda
    d_list = np.array([-0.1, -0.01, -0.001, 0, 0.001, 0.01, 0.1])
    
    # Generisanje stohastičkog koraka
    dx_list = d_list*random.random()
    dy_list = d_list*random.random()
    
    # Nasumičan izbor koraka
    dx = random.choice(dx_list)
    dy = random.choice(dy_list)
    
    # Dobijanje suseda uz zadovoljenje ograničenja opsega
    xp = max(-1,min(1, x + dx))
    yp = max(-1,min(1, y + dy))
    
    return xp, yp
```

### Primer generisanja suseda

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))
ax = axes.ravel()

plt.suptitle("Generisanje suseda")
for j in range(4):
    # Nasumičan izbor inicijalne pozicije
    x = random.random()
    y = random.random()

    # Broj suseda
    N = 20
    # Niz suseda
    susedi = np.ones((N,2), dtype=float)

    for i in range(N):
        susedi[i,:] = generisanje_suseda(x,y)

    ax[j].scatter(susedi[:,0], susedi[:,1])
    ax[j].scatter(x, y)
    ax[j].set_xlabel("x")
    ax[j].set_ylabel("y")
    ax[j].set_title("Inicijalna pozicija (x,y)=(%.3f,%.3f)" %(x,y))
    ax[j].legend(["Pozicija suseda", "Početna pozicija"])
    ax[j].grid('minor')

plt.show()
fig.savefig('C:\Z\ETF\Vestacka inteligencija\DZ1\Izvestaj\generisanje_suseda.png', bbox_inches='tight')
```

# Simulirano kaljenje

## Definicija funkcije za simulirano kaljenje

```python
def simulirano_kaljenje(x, y, T, Mk, mod=False, show_steps=False):
    """
    Opis funkcije:
        Vrši se maksimizacija ciljne funkcije metodom
        simuliranog kaljenja. Počev od zadatih početnih
        stanja generišu se potencijalna rešenja (xp)
        i računa se razlika ciljne funkcije novog i
        trenutnog rešenja. Ukoliko je rešenje bolje,
        ono se zadržava dok se u slučaju lošijeg rešenja
        ono zadržava sa verovatnoćom koja eksponencijalno
        opada sa temperaturom kaljenja.
    Ulazni argumenti:
        x : Koordinata početnog stanja
        y : Koordinata početnog stanja
        T : Vektor rasporeda hlađenja
        Mk : Broj ponavljanja algoritma na Tk
        mod : Da li se koristi modifikovani algoritam
    Izlazni argumenti:
        fopt : Optimalna vrednost ciljne funkcije
        iopt : Broj iteracije u kojoj je dostignuta optimalna vrednost
        f_values : Vrednosti ciljne funkcije po iteracijama
        xval : x koordinate rešenja po iteracijama
        yval : y koordinate rešenja po iteracijama
        
    """
    # Provera vrednosti početnih stanja
    if((x<-1) or (x>1)):
        raise ValueError("Argument x je van očekivanog opsega [-1,1]")
    elif((y<-1) or (y>1)):
        raise ValueError("Argument y je van očekivanog opsega [-1,1]")
    
    # Najbolje rešenje
    fopt = fc(x, y)
    xopt = x
    yopt = y
    iopt = 1
    xval = np.zeros((len(T)*Mk,1), dtype=float)
    yval = np.zeros((len(T)*Mk,1), dtype=float)
    
    # Lista vrednosti ciljne funkcije u svakoj od iteracija
    f_values = []
    
    # Brojač trenutne iteracije
    itr = 0
    
    # Lista verovatnoća prihvatanja lošeg rešenja
    probal = []
    # Lista grešaka
    errorl = []
    # Lista temperatura
    Tl = []
    # Niz usrednjenih verovatnoća prihvatanja lošijeg rešenja
    probam = np.zeros((len(T),2), float)
    
    
    # Prolazak kroz temperature prosleđene rasporedom hlađenja
    for it in range(len(T)):
        # Trenutna temperatura
        Tk = T[it]
        
        # Pomoćne liste
        probal_ = []
        
        # Ponavljanje algoritma na temperaturi Tk
        for m in range(Mk):
            # Broj trenutne iteracije
            itr += 1
            
            # Generisanje novog rešenja modifikacijom trenutnih x i y
            xp, yp = generisanje_suseda(x, y)
            
            # Izračunavanje ciljnih funkcija
            f = fc(x, y)
            fp = fc(xp, yp)
            
            # Razlika ciljnih funkcija
            delta = fp - f
            
            # Provera razlike ciljne funkcije novog i trenutnog rešenja
            if(delta > 0):
                # Ciljna funkcija se povećala, rešenje se zadržava
                x = xp
                y = yp
            else:
                # Ciljna funkcija se smanjuje
                if(Tk==0):
                    # Izbegavanje deljenja sa nulom
                    x = x
                    y = y

                elif(random.random() < np.exp(delta/Tk)):
                    x = xp
                    y = yp
                    
                # Čuvanje vrednosti
                if(show_steps):
                    if(Tk==0):
                        probal_.append(0)
                    else:
                        probal_.append(np.exp(delta/Tk))

                    Tl.append(Tk)
                    errorl.append(delta)

            # Preračunavanje ciljne funkcije
            f = fc(x, y)
                  
            # Provera najboljeg dosadašnjeg rešenja
            if((f - fopt)>1e-7):
                xopt = x
                yopt = y
                fopt = f
                iopt = itr
            
            # Modifikovani algoritam
            if (mod and (it*Mk+m)>20):
                # Ukoliko je srednja vrednost ciljne funkcije u poslednjih
                # nekoliko iteracija mala, algoritam se zaglavio i položaj
                # se reinicijalizuje
                if(abs(np.var(f_values[-20:]))<0.00001):
                    x = xopt
                    y = yopt
                
            # Pamćenje vrednosti ciljne funkcije za svaku iteraciju
            f_values.append(f)
                
            # Čuvanje trenutnih rešenja
            xval[it*Mk+m] = x
            yval[it*Mk+m] = y
            
        if(show_steps):
            probal += probal_
            probam[it, :] = [np.mean(probal_), len(probal)-len(probal_)/2]
        
            
    if(show_steps):
        fig = plt.figure(figsize=(12,8))
        plt.plot(probal)
        plt.plot(Tl)
        plt.plot(errorl)
        plt.plot(probam[:,1],probam[:,0],c='#FF0000',marker='x',alpha=1)
        plt.title("Zavisnost prihvatanja lošijeg rešenja od opsega temperature (Tmax = %.2f)"%(np.max(T)))
        plt.xlabel('Broj iteracije')
        plt.legend(["Verovatnoća prihvatanja lošijeg rešenja",
                   "Temperatura",
                   "Razlika ciljne funkcije",
                   "Usrednjena verovatnoća"], loc='upper right')
        plt.show()
        fig.savefig('C:\Z\ETF\Vestacka inteligencija\DZ1\Izvestaj\T_opseg_%.2f.png'%(np.max(T)), bbox_inches='tight')
            
                
    return fopt, iopt, f_values, xval, yval
```

## Analiza uticaja izbora maksimalne temperature

```python
for Tmax in [0.01, 0.1, 10]:
    
    # Raspored hlađenja
    T = np.flip(np.logspace(0,4,50))
    T = (T - np.min(T))/(np.max(T) - np.min(T))*Tmax
    T = T.tolist()

    # Inicijalna stanja
    x_init = random.random()*2-1
    y_init = random.random()*2-1
    
    # Broj iteracija
    Mk = 100
    
    _ = simulirano_kaljenje(x_init, y_init, T, Mk, show_steps=True)
```

## Simulacija eksperimenta

### Funkcija za simulaciju eksperimenta

```python
def simulacija_SK(x, y, T, Mk, mod):
    """
    Opis funkcije:
        Funkcija služi za smanjenje obima koda za simuliranje ponašanja
        algoritma sa i bez modifikacije. Definisani su rasporedi hlađenja
        i početna stanja algoritma za 100 realizacija. Dobijene vrednosti
        su usrednjene po realizacijama i zapamćene su maksimalne vrednosti
        funkcije kao i njeni argumenti.
    Ulazni argumenti:
        x : x koordinate početnih rešenja
        y : y koordinate početnih rešenja
        T : Lista rasporeda hlađenja
        Mk : Broj iteracija pri fiksnoj temperaturi
        mod : Modifikacija algoritma
    Izlazni argumenti:
        fmean: Usrednjene vrednosti ciljnih funkcija po iteracijama
        fmax: Optimalne vrednosti ciljnih funkcija po rasporedu hlađenja
        xmax: Optimalne vrednosti x koordinate po rasporedu hlađenja
        ymax: Optimalne vrednosti y koordinate po rasporedu hlađenja
        imax: Iteracija u kojoj je dostignuta optimalna vrednost
    """
    # Broj realizacija
    Nr = x_init.size
    
    # Broj rasporeda hlađenja
    Nt = len(T)
    
    # Promenljive za skladištenje rezultata
    fmean = []
    fmax = np.zeros(Nt, dtype=float)
    iopt_r = np.zeros((Nt,Nr), dtype=int)
    fopt_r = np.zeros((Nt,Nr), dtype=float)

    # Simulacija
    for it in range(len(T)):
        
        # Matrica za smeštanje vrednosti ciljne funkcije po iteracijama
        fmean_ = np.zeros((Nr, len(T[it])*Mk))
        
        # Prolazak kroz realizacije za različite početne pozicije
        for ir in range(Nr):
            
            # Ispis trenutne iteracije
            if (ir != (Nr-1)):
                print("%s: broj realizacije %d/%d" %(titles[it],ir+1,Nr), end='\r')
            else:
                print("%s: broj realizacije %d/%d" %(titles[it],ir+1,Nr), end='\n')
            
            # Pozivanje funkcije za simulirano kaljenje
            fopt, iopt, f_values = simulirano_kaljenje(x[ir], y[ir], T[it], Mk, mod)[:3]
            
            # Pamćenje promene ciljne funkcije po iteracijama
            fmean_[ir, :] = np.asarray(f_values).flatten()
            
            # Pamćenje najbolje vrednosti ciljne funkcije
            if (fopt > fmax[it]):
                fmax[it] = fopt
            
            # Maksimalna vrednost ciljne funkcije trenutne realizacije
            fopt_r[it,ir] = fopt
            # Broj iteracije u kojoj je dostugnuta maksimalna vrednost
            iopt_r[it,ir] = iopt
                
        # Usrednjavanje realizacija i čuvanje u listu
        fmean.append(np.mean(fmean_, axis=0))
    
    return fmean, fmax, fopt_r, iopt_r
```

# Pretraga po snopu

## Definicija funkcije pretrage po snopu

```python
def pretraga_po_snopu(x, y, Np, Nk, Ni):
    """
    Opis funkcije:
        U svakoj iteraciji svaka jedinka generiše Nk kandidata
        za narednu populaciju. Na osnovu Np*Nk kandidata bira
        se Np kandidata koji imaju najveću vrednost ciljne f-je.
        Na kraju svake iteracije vrši se provera da li je najbolji
        kandidat (sa najvećom vrednošću ciljne funkcije) postigao
        najbolji rezultat zabeležen do tada. Ukolko je uslov ispunjen
        algoritam nastavlja sa pretragom, u suprotnom se algoritam
        zaustavlja.
        
    Ulazni argumenti:
        x : x koordinate početnih rešenja
        y : y koordinate početnih rešenja
        Np : Veličina populacije
        Nk : Broj kandidata jedne jedinke
        Ni : Broj iteracija
    Izlazni argumenti:
        fopt : Optimalna vrednost ciljne funkcije
        iopt : Iteracija u kojoj je dostignuta optimalna vrednost
        fvalues : Vrednost ciljne funkcije po iteracijama
        xval : x koordinate populacije rešenja po generacijama
        yval : y koordinate populacije rešenja po generacijama
    """
    
    # Brojač iteracija
    itr = 0
    # Najbolja vrednost ciljne funkcije
    fopt = -5
    # Promenljiva za čuvanje vrednosti ciljne funkcije kroz populacije
    fvalues = []
    # Pozicije populacije kroz generacije
    xval = np.zeros((Ni, Np), dtype=float)
    yval = np.zeros((Ni, Np), dtype=float)
    
    for i in range(Ni):
        # Struktura za čuvanje vrednosti ciljne funkcije kandidata
        fk = {}
        
        # Niz za skladištenje koordinata
        xk = np.zeros((Np*Nk,1), dtype=float)
        yk = np.zeros((Np*Nk,1), dtype=float)
        
        # Za svaku jedinku u populaciji generišu se kandidati
        for j in range(Np):
            for s in range(Nk):
                # Generisanje suseda
                xs, ys = generisanje_suseda(x[j],y[j])
                
                # Proračun ciljne funkcije
                fs = float(fc(xs, ys))

                # Skladištenje vrednosti
                br_kandidata = j*Nk+s
                fk[br_kandidata] = fs
                xk[br_kandidata] = xs
                yk[br_kandidata] = ys
                
        # Sortiranje kandidata po vrednosti ciljne funkcije (DESC)
        ind = sorted(fk, key=fk.get, reverse=True)

        # Najboljih Np kandidata čini novu populaciju
        x = xk[ind[:Np]]
        y = yk[ind[:Np]]
        
        # Pamćenje pozicija
        xval[i,:] = x.ravel()
        yval[i,:] = y.ravel()
        
        # Vrednosti ciljne funkcije
        f_kandidata = np.array(list(fk.values()))

        # Dodavanje usrednje vrednosti ciljne funkcije u listu
        fvalues.append(np.mean(f_kandidata[ind[:Np]]))

        # Provera najboljeg rešenja
        if((fk[ind[0]] - fopt) > 0.000001):
            fopt = fk[ind[0]]
            iopt = i
            
    return fopt, iopt, fvalues, xval, yval
```

## Kodiranje rešenja

### Funkcije za kodiranje/konverziju

```python
def kodiranje(x, N=32):
    """
    Opis funkcije:
        Kako je posmatrani opseg ulaznih argumenata ciljne funkcije
        u opsegu [-1,1] dovoljno je razmatrati opseg (-1,1) tako da
        opštost algoritma, kao i njegova optimalnost i kompletnost
        nisu ugroženi. Modikovani binarni kod se sastoji od N cifara
        i prva cifra predstavlja znak ulaznog broja, dok naredna N-1
        cifra predstavlja binarni zapis nakon decimalnog zareza.
    Ulazni argumenti:
        x : Broj koji treba konvertovati
        N : Duzina decimalnog zapisa broja
    Izlazni argumenti:
        binarni_zapis : Binarni zapis broja x
    """
    if(x >= 1 or x <= -1):
        raise ValueError('Podrazumevani opseg ulazne promenljive je (-1,1)')
    
    # Prva cifra modifikovanog binarnog zapisa
    if(x < 0):
        celi_deo_bin = '1'
    else:
        celi_deo_bin = '0'

    raz_deo = ('{:f}'.format(round(x, N))).split('.')[1]

    # Moduo broja korišćen za konverziju
    mod = 10**(len(raz_deo))
    
    # Konverzija u celobrojnu promenljivu
    raz_deo = int(raz_deo)
    
    # Binarni zapis razlomljenog dela
    raz_deo_bin = ''
    for i in range(N):
        raz_deo *= 2
        if(raz_deo >= mod):
            raz_deo_bin += '1'
            raz_deo %= mod
        else:
            raz_deo_bin += '0'
    
    binarni_zapis = ''.join([celi_deo_bin, raz_deo_bin])
    return binarni_zapis
    
def dekodiranje(x):
    """
    Opis funkcije:
        Funkcija je inverzna funkciji kodiranje i njom
        se prelazi iz binarnog zapisa, u kojem prva cifra
        predstavlja znak broja dok ostale predstavljaju
        binarni zapis iza decimalnog zareza, u decimalni
        zarez množenjem cifara binarnog zapisa stepenom
        broja 2.
    Ulazni argumenti:
        x : Broj koji treba konvertovati
    Izlazni argumenti:
        num : Decimalni zapis broja x
    """
    
    # Decimalni zapis broja
    num = 0
    
    # Faktor (stepen broja 2)
    faktor = 1
    for c in x[1::]:
        # Umanjenje faktora
        faktor /= 2
        
        # Konverzija u decimalni zapis
        num += int(c)*faktor
        
    # Konverzija znaka
    if(x[0] == '1'):
        num = -num
        
    return num
```

## Selekcija

### Funkcija za selekciju

```python
def selekcija(x, y, Np, mod='norm', show_steps=False):
    """
    Opis funkcije:
        Na osnovu prosleđenih parova koordinata i njihovih vrednosti ciljnih
        funkcija vrši se selekcija jedinki. Parametar 'mod' određuje tip
        selekcije, metoda elitizma ili izbor jedinke sa verovatnoćom koja je
        proporcionalna normalizovanoj vrednosti ciljne funkcije.
    Ulazni argumenti:
        x : Argument ciljne funkcije
        y : Argument ciljne funkcije
        Np : Veličina populacije
    Izlazni argumenti:
        xs : x koordinata selektovanih jedinki
        ys : y koordinata selektovanih jedinki
    """
    
    # Broj jedinki
    N = x.size
    
    # Niz vrednosti ciljne funkcije  
    fval = fc(x,y)
    
    # Normalizacija vrednosti ciljne funkcije
    fval /= np.sum(fval)
    
    # Opcioni prikaz međukoraka
    if(show_steps):
        print("Normalizovane vrednosti ciljne funkcije:")
        print(fval.flatten())

    
    if mod == 'elit':
        ########################## Metoda elitizma ###########################
        
        # Indeksi sortiranih vrednosti
        ind = fval.flatten().argsort()
        
        # Indeksi selektivanih jedinki
        ind_sel = ind[-Np:]
        
        if(show_steps):
            print("Selektovani indeksi:")
            print(ind_sel)
        
    else:
        ##### Verovatnoća izbora proporcionalna vrednosti ciljne funkcije #####
        
        # Indeksi jedinki
        ind = np.arange(N)
        
        # Indeksi izabranih jedinki
        ind_sel = random.choices(ind, weights=fval.flatten(), k=Np)
        

    # Nizovi za smeštanje selektovanih jedinki
    xs = x[ind_sel].reshape((Np,1))
    ys = y[ind_sel].reshape((Np,1))

    
    return xs, ys 
```

## Ukrštanje

### Funkcija za ukrštanje

```python
def ukrstanje(x, y, weights):
    """
    Opis funkcije:
        Ukrštanje se vrši na dva slučajno odabrana mesta.
        U slučaju da su izabrana mesta identična, vrši
        se ukrštanje u jednoj tački na izabranom mestu.
    Ulazni argumenti:
        x : Broj u binarnom zapisu
        y : Broj u binarnom zapisu
        weights : Težine izbora mesta ukrštanja
    Izlazni argumenti:
        xy : Rezultat ukrštanja
        yx : Rezultat ukrštanja
        mesto : Mesta ukrštanja
    """
    
    # Dužina binarnog zapisa
    N = len(x)
    
    
    # Izbor mesta ukrštanja
    mesto = random.choices(np.arange(N), weights=weights, k=2)[:2]
    mesto.sort()
    
    if (mesto[0] != mesto[1]):
        # Ukrštanje na 2 mesta
        xy = ''.join([x[:mesto[0]], y[mesto[0]:mesto[1]], x[mesto[1]:]])
        yx = ''.join([y[:mesto[0]], x[mesto[0]:mesto[1]], y[mesto[1]:]])
    else:
        # Ukrštanje na jednom mestu
        xy = ''.join([x[:mesto[0]], y[mesto[0]:]])
        yx = ''.join([y[:mesto[0]], x[mesto[0]:]])
    
    return xy, yx, mesto
```

## Mutacija

### Funkcija za mutaciju

```python
def mutacija(x, p, weights):
    """
    Opis funkcije:
        Generisanjem slučajnog broja na osnovu verovatnoće p
        se određuje da li će doći do mutacije. Ukoliko je taj
        uslov ispunjen, do mutacije dolazi tako što se na
        slučajan način bira mesto mutacije (u opsegu dužine x)
        i vrednost na mestu te cifre (0 ili 1).
    Ulazni argumenti:
        x : Broj u binarnom zapisu
        p : Verovatnoća mutacije
    Izlazni argumenti:
        xm : Broj nakon mutacije
        info : Informacije o mutaciji
    """
    
    # Dužina binarnog zapisa
    N = len(x)
    
    # Mutirani broj
    xm = list(deepcopy(x))
    
    isM = False # Indikator mutacije
    mesto = -1
    vrednost = -1
    
    # Slučajan događaj kojim se određuje pojava mutacije
    if (random.random() < p):
        isM = True
        
        # Odabir mesta mutacije
        mesto = random.choices(np.arange(0,N), weights=weights, k=1)[0]
        
        # Odabir vrednosti mutacije
        if(xm[mesto] == '1'):
            vrednost = '0'
        else:
            vrednost = '1'
        
        # Mutacija
        xm[mesto] = vrednost
        
    # Pakovanje informacija o mutaciji
    info = [isM, mesto, vrednost]
        
    return ''.join(xm), info
```

## Genetski algoritam - funkcija

```python
def genetski_algoritam(x, y, Np, Ng, p, Nk=16):
    """
    Opis funkcije:
        Kroz svaku generaciju trenutna rešenja prolaze kroz proces
        selekcije, zatim se kodiraju i vrši se njihovo ukrštanje i 
        mutacija. Nakon prethodno završenih koraka, obračunavaju se
        vrednosti ciljne funkcije i pamti najbolje rešenje.
    Ulazni argumenti:
        x : Niz početnih x koordinata
        z : Niz početnih y koordinata
        Np : Broj jediniki u jednoj populaciji
        Ng : Broj generacija
        p : Verovatnoća mutacije
        Nk : Dužina kodirane vrednosti u binarnom zapisu
    Izlazni argumenti:
        fopt : Maksimalna vrednost funkcije
        ig_opt : Iteracija u kojoj je dostignuto rešenje
        fun_gen : Lista vrednosti ciljne funkcije po generacijama
        xval : x koordinate jedinki kroz populacije
        yval : y koordinate jedinki kroz populacije
    """
    # Optimalne vrednosti
    fopt = 0
    ig_opt = -1
    
    # Vrednosti ciljne funkcije po generacijama
    fun_gen = []
    xval = []
    yval = []
    
    # Težine mesta mutacije
    weights_mut = np.ones(Nk+1)
    weights_mut[:4] = 0
    
    # Težine mesta za ukrštanje
    weights_ukr = np.zeros(Nk+1)
    weights_ukr[2:] = 1
    
    #weights_ukr[2:] = 1
    for ig in range(Ng):        
        
        # Selekcija
        xs, ys = selekcija(x, y, Np)

        
        # Kodiranje
        xk = [kodiranje(float(xs[i]), Nk) for i in range(Np)]
        yk = [kodiranje(float(ys[i]), Nk) for i in range(Np)]
        
        
        
        # Mešanje jedinki pred ukrštanje
        k = list(zip(xk, yk))
        random.shuffle(k)
        xk, yk = zip(*k)
        xk = list(xk)
        yk = list(yk)

        # Ukrštanje
        xu = [ukrstanje(xk[i], xk[i+1], weights_ukr)[:2] for i in range(0,Np-1,2)]
        yu = [ukrstanje(yk[i], yk[i+1], weights_ukr)[:2] for i in range(0,Np-1,2)]
        
        # Konverzija u listu
        xu = [xii for xi in xu for xii in xi]
        yu = [yii for yi in yu for yii in yi]
        
        
        
        # Mutacija
        xm = [mutacija(xi, p, weights_mut)[0] for xi in xu]
        ym = [mutacija(yi, p, weights_mut)[0] for yi in yu]

        
        
        # Dekodiranje
        xf = [dekodiranje(str(xi)) for xi in xm]
        yf = [dekodiranje(str(yi)) for yi in ym]
        
        
        
        # Rešavanje ugnježdenih listi i konverzija u niz
        xs = np.array([xss[0] for xss in xs]).flatten()
        ys = np.array([yss[0] for yss in ys]).flatten()
        xf = np.array(xf).flatten()
        yf = np.array(yf).flatten()
        
        # Selekcija jedinki iz stare i nove generacije metodom elitizma
        xs, ys = selekcija(xs, ys, 2*Np//10, mod='elit')
        xf, yf = selekcija(xf, yf, 8*Np//10, mod='elit')
        
        # Spajanje 'elitnih' jedinki iz stare i nove generacije
        xf = np.concatenate((xs, xf), axis=0)
        yf = np.concatenate((ys, yf), axis=0)
        

        # Čuvanje pozicija jedinki
        xval.append(xf)
        yval.append(yf)
        
        # Izračunavanje ciljne funkcije
        fun = [fc(xi, yi) for xi, yi in zip(xf, yf)]

        # Usrednjavanje po populaciji
        f_mean = np.mean(fun)
        
        # Maksimalna vrednost ciljne funkcije u populaciji
        fmax_pop = np.amax(fun)
        
        # Provera optimalnog rešenja
        if((fmax_pop - fopt) > 1e-7):
            fopt = fmax_pop
            ig_opt = ig
                
        # Uprosečena vrednost ciljne funkcije generacije
        fun_gen.append(f_mean)
        
        # Nova populacija
        x, y = np.array(xf).reshape((Np,1)), np.array(yf).reshape((Np,1))
                
    return fopt, ig_opt, fun_gen, xval, yval
```

## Simulacija algoritama

```python
# Broj realizacija
Nr = 100

# Promeljiva za čuvanje iteracija u kojima se dolatzi do konačnog rešenja
iteracije = np.zeros((3, Nr), dtype=int) 
# Promenljiva za čuvanje najboljih vrednosti ciljnih funkcija
f_vrednosti = np.zeros((3, Nr), dtype=float)

for i in range(Nr):
    
    ###################################################################
    ###################### SIMULIRANO KALJENJE ########################
    ###################################################################
    
    # Maksimalna temperatura hlađenja
    Tmax = 0.1

    # Optimalno hlađenje
    T = np.flip(np.logspace(0, 4, 50))
    T = (T - np.min(T))/(np.max(T) - np.min(T))*Tmax

    # Generisanje početnh rešenja
    x_init = random.random()*2-1
    y_init = random.random()*2-1

    # Broj iteracija za jednu temperaturu
    Mk = 100
    
    # Poziv algoritma
    f, itr = simulirano_kaljenje(x_init, y_init, T, Mk, mod=True, show_steps=False)[:2]
                 
    # Čuvanje vrednosti
    f_vrednosti[0,i] = f
    iteracije[0,i] = itr
        
        
    ###################################################################
    ####################### PRETRAGA PO SNOPU #########################
    ###################################################################
    
    
    # Broj jedinki u populaciji
    Np = 40
    # Broj iteracija/generacija
    Ni = 8
    # Broj generisanih kandaidata jedne jedinke
    Nk = 20

    # Inicijalna populacija
    x_init = np.array([random.random()*2-1 for i in range(Np)]).reshape((Np,1))
    y_init = np.array([random.random()*2-1 for i in range(Np)]).reshape((Np,1))
        
    # Poziv funkcije pretrage po snopu
    f, itr = pretraga_po_snopu(x_init, y_init, Np, Nk, Ni)[:2]
    
    # Čuvanje vrednosti
    f_vrednosti[1,i] = f
    iteracije[1,i] = itr*Np*Nk
    
    
    ###################################################################
    ####################### GENENTSKI ALGORITAM #######################
    ###################################################################
    
    
    # Veličina populacije
    Np = 100
    # Broj generacija
    Ng = 40
    # Verovatnoća mutacije
    p = 0.05

    # Inicijalna populacija
    x_init = np.array([random.uniform(-1,1) for i in range(Np)]).reshape((Np,1))
    y_init = np.array([random.uniform(-1,1) for i in range(Np)]).reshape((Np,1))

    # Poziv funkcije genetskog algoritma
    f, itr = genetski_algoritam(x_init, y_init, Np, Ng, p, Nk=12)[:2]
    
    # Čuvanje vrednosti
    f_vrednosti[2,i] = f
    iteracije[2,i] = itr*Np
    
    
    # Ispis trenutne iteracije
    if (i != (Nr-1)):
        print("Broj realizacije %d/%d" %(i+1,Nr), end='\r')
    else:
        print("Broj realizacije %d/%d" %(i+1,Nr), end='\n')
```
